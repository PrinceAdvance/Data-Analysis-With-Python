```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#naming the filepath
movie_filepath = "C:/Users/Admin/Downloads/Movie project.xlsx"
```

```python
movie_data = pd.read_excel(movie_filepath)
```

```python
#Checking if the data was loaded correctly by loading the whole data 
movie_data
```

```python
domestic_percentage = (movie_data['Domestic']/movie_data['Worldwide']*100).round(1)
foreign_percentage = (movie_data['Foreign']/movie_data['Worldwide'] *100).round(1)

new_movie = pd.DataFrame(
    {"Release Group": movie_data["Release Group"],
     "Budget": movie_data["Budget"],
     "Revenue worldwide": movie_data["Worldwide"],
     "Domestic revenue": movie_data["Domestic"],
     "Domestic percentage(%)": domestic_percentage,
     "Foreign revenue": movie_data["Foreign"],
     "Foreign percentage(%)": foreign_percentage}
)
```

```python
new_movie
```

```python
#Checking the data types to make sure they are all in the right format,NB:If budget and worldwide columns were object types calculation will be difficult
new_movie.dtypes
```

```python
#Describing the data statistically
new_movie.describe()
```

```python
new_movie.info()
```

```python
#Check for missing values,data looks squeky clean!
new_movie.isnull().sum()
```

```python
# 
movie_profit = new_movie['Revenue worldwide'] - new_movie['Budget']
```

```python
movie_profit
```

```python
#Plotting a chart to show the cost of production against worldwide box office for each movie
plt.figure(figsize=(10, 10))
new_movie.plot(x="Release Group", y=["Revenue worldwide", "Budget"], kind="bar")
plt.title("A chart showing the cost of production against worldwide box office for each movie")
plt.xticks(rotation=45, ha='right')
```

```python
#Plotting a bar chart to show how movies fared worldwide and domestically 
new_movie.plot(x="Release Group", y=["Domestic revenue", "Foreign revenue"], kind="bar")
```

```python
# Creating a DataFrame with the Release Group and the corresponding profits
data = pd.DataFrame({"Release Group": new_movie["Release Group"], "Profit": movie_profit})

# Sort the DataFrame by the 'Profit' column in descending order
data_sorted = data.sort_values(by="Profit", ascending=False)

# Plot the sorted bar chart
plt.bar(data_sorted["Release Group"], data_sorted["Profit"])
plt.xticks(rotation=45, ha='right')
plt.title("A Chart Showing the Profit Each Movie Made")
plt.xlabel("Release Group")
plt.ylabel("Profit")
plt.show()
```

```python
# Assuming new_movie is your DataFrame with columns 'Budget', 'Revenue worldwide', and 'Release Group'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=new_movie, x="Budget", y="Revenue worldwide", hue="Release Group", palette="viridis", s=100, alpha=0.7)

# Adjust plot aesthetics
plt.title("Scatter Plot of Budget vs Worldwide Revenue with Release Group")
plt.xlabel("Budget")
plt.ylabel("Revenue worldwide")
plt.legend(title="Release Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=new_movie, x="Budget", y= movie_profit, hue="Release Group", palette="viridis", s=100, alpha=0.7)

# Adjust plot aesthetics
plt.title("Scatter Plot of Budget vs Profit with Release Group")
plt.xlabel("Budget")
plt.ylabel("Revenue worldwide")
plt.legend(title="Release Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
